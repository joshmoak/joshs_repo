## Scraping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
pd.set_option('display.max_rows', 200)
```

```python
# urls = pd.read_csv('appic urls.csv')
urls = pd.read_csv('appic urls.csv')['URL'].to_list()
```

```python
# Now that we have a url, how to parse tables and info
# url = 'https://membership.appic.org/directory/display/741'
# url = 'https://membership.appic.org/directory/display/376'
url = urls[25]
print(url)
```

```python
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')

tables = soup.find_all('table')
features = []
values = []

for i, table in enumerate(tables[:18]):
    print('+'*50 + f' TABLE {i}', '+'*50)
    for row in table.find_all('tr'):
        th = row.find('th')
        td = row.find('td')
        if td and th and (th.get('scope') == 'row' or th.get('scope') is None):
            print(th.text.strip(), td.text.strip())
            print('-'*50)
print('+'*50 + ' PROGRAM DESCRIPTION', '+'*50)
print(soup.find('div', class_ = 'program-desc').text)
# print('-'*100)

# TODO check table 11 for repeated info
```

```python
if torch.cuda.is_available():
    device = torch.device("cuda")
    total_memory = torch.cuda.get_device_properties(device).total_memory / (1024**2) # bytes to MiB
    allocated_memory = torch.cuda.memory_allocated(device) / (1024**2)
    reserved_memory = torch.cuda.memory_reserved(device) / (1024**2)
    available_memory = total_memory - allocated_memory

    print(f"GPU Total Memory: {total_memory:.2f} MiB")
    print(f"GPU Allocated Memory (by PyTorch): {allocated_memory:.2f} MiB")
    print(f"GPU Reserved Memory (by PyTorch): {reserved_memory:.2f} MiB")
    print(f"GPU Effectively Available Memory: {available_memory:.2f} MiB")

    # To free up unused reserved memory
    # torch.cuda.empty_cache()
else:
    print("CUDA not available")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

model_name = "meta-llama/Llama-2-7b-chat-hf"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",       # automatically splits layers between GPU/CPU
    torch_dtype="auto"       # auto selects FP16 or BF16
)

generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
```
