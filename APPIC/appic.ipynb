```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
pd.set_option('display.max_rows', 200)
```

```python
# urls = pd.read_csv('appic urls.csv')
urls = pd.read_csv('joshs_repo/appic urls.csv')['URL'].to_list()
```

```python
# Now that we have a url, how to parse tables and info
# url = 'https://membership.appic.org/directory/display/741'
# url = 'https://membership.appic.org/directory/display/376'
for url in urls[0:15]:
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    print(soup.find('div', class_ = 'program-desc').text)
    print('-'*100)
```

```python
tables = soup.find_all('table')
# df = pd.DataFrame()
features = []
values = []

for i, table in enumerate(tables[:18]):

    for row in table.find_all('tr'):
        th = row.find('th')
        td = row.find('td')
        if td and th and (th.get('scope') == 'row' or th.get('scope') is None):
            # df[th.text.strip()] = [td.text.strip()]
            # features.append(th.text.strip())
            print(th.text.strip(), td.text.strip())
            print('-'*50)

    print(i, '+'*100)

# Table 2: just grab the training director. might just leave it as is
# Table 11: weird nested tables on /741, but not on 376
```
