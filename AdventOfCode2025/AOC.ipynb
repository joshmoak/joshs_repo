```python
with open('input/aoc2.txt') as f:
    ranges = f.read().split(',')
```

```python
print(ranges[0:5])
print(len(ranges))
```

```python
# Part 1
s = []
for r in ranges:
    start, end = r.split('-')

    nums = [str(num) for num in range(int(start), int(end)+1) if len(str(num))%2 == 0]

    first_halves = list(set([num[0:int(len(num)/2)] for num in nums]))
    second_halves = list(set([num[int(len(num)/2):] for num in nums]))

    for num in first_halves:
        if num in second_halves and int(num+num) <= int(end) and int(num+num) >= int(start):
            s.append(int(num+num))
sum(s)

```

```python
# Part 2
import re
pattern = re.compile(r"^(.+)\1+$")
bool(pattern.match('123123123'))
s = []
for r in ranges:
    start, end = r.split('-')

    nums = [str(num) for num in range(int(start), int(end)+1)]
    for num in nums:
        if bool(pattern.match(num)):
            s.append(int(num))

sum(s)
```

```python
from itertools import combinations
import numpy as np

with open('input/aoc3.txt') as f:
    banks = f.read().rstrip().split('\n')
banks[0:5]
```

```python
# Part 1
s = []
for b in banks:
    s.append(max([int(''.join(c)) for c in combinations(b, r = 2)]))
sum(s)
```

```python
# Part 2
s = []
for bank in banks:
    start = 0
    joltages = []
    bank = [int(i) for i in bank]
    for i in reversed(range(12)):
        if i == 0:
            search_space = bank[start:]
        else:
            search_space = bank[start:-i]
        start += np.argmax(search_space) + 1
        joltages.append(str(max(search_space)))
    s.append(int(''.join(joltages)))
sum(s)
```

```python
with open('input/aoc4.txt') as f:
    grid = []
    for line in f.read().rstrip().split():
        grid.append(list(line))
import numpy as np
grid = np.array(grid)
grid
```

```python
grid.shape
```

```python
# Part 1
can_access = []
for i in range(grid.shape[0]):
    for j in range(grid.shape[1]):
        if grid[i,j] == '@':
            num_adj = 0
            for spot in [(i-1, j-1), (i-1, j), (i-1, j+1), (i, j-1), (i, j+1), (i+1, j-1), (i+1, j), (i+1, j+1)]:
                if spot[0] >= 0 and spot[0] < grid.shape[0] and spot[1] >= 0 and spot[1] < grid.shape[1] and grid[spot] == '@':
                    num_adj += 1
            if num_adj < 4:
                can_access.append((i,j))
len(can_access)
```

```python
# Part 2
stuck = False
num_removed = 0
while not stuck:
    can_access = []
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            if grid[i,j] == '@':
                num_adj = 0
                for spot in [(i-1, j-1), (i-1, j), (i-1, j+1), (i, j-1), (i, j+1), (i+1, j-1), (i+1, j), (i+1, j+1)]:
                    if spot[0] >= 0 and spot[0] < grid.shape[0] and spot[1] >= 0 and spot[1] < grid.shape[1] and grid[spot] == '@':
                        num_adj += 1
                if num_adj < 4:
                    can_access.append((i,j))
    num_removed += len(can_access)
    if len(can_access) == 0:
        stuck = True
    for pair in can_access:
        grid[pair] = '.'
num_removed
```

```python
with open('input/aoc5.txt') as f:
    lines = f.read().rstrip().split('\n\n')
    ranges = [(int(range.split('-')[0]), int(range.split('-')[1])) for range in lines[0].split()]
    IDs = [int(ID) for ID in lines[1].split()]
```

```python
len(ranges), len(IDs)
```

```python
# Part 1
fresh = []
for ID in IDs:
    for r0, r1 in ranges:
        if r0 <= ID <= r1:
            fresh.append(ID)
print(len(set(fresh)))
```

```python
# Part 2:
def merge(to_merge):
    merged_ranges = []
    for i, r in enumerate(to_merge):
        if i < len(to_merge) - 1:
            nr = to_merge[i+1]
            if r[1] < nr[0]: # no overlap
                merged_ranges.append(r)
            elif r == nr:
                continue
            elif r[1] >= nr[0]: # Overlap
                if r[1] >= nr[1]:
                    merged_ranges.append(r)
                if r[1] < nr[1]:
                    merged_ranges.append((r[0], nr[1]))
                to_merge.remove(nr)   
    merged_ranges.append(to_merge[-1])
    return merged_ranges
    
sorted_ranges = sorted(ranges)
merging = True
step = sorted_ranges
while merging:
    next_step = merge(step.copy())
    if step == next_step:
        merging = False
    step = next_step
    
s = 0
for r in step:
    s += 1 + r[1] - r[0]
print(s)
```

```python
import numpy as np
data = np.loadtxt('input/aoc6.txt', dtype = str)
numbers = data[0:-1].astype(float).T
operations = data[-1]
numbers, operations
```

```python
# Part 1
total = 0
for operation, row in zip(operations, numbers):
    if operation == '*':
        total += row.prod()
    if operation == '+':
        total += row.sum()

int(total)
```

```python
# Part 2
with open('input/aoc6.txt') as f:
    lines = f.read().split('\n')
ops = lines[-1]
```

```python
ops_clean = []
for i in range(len(ops)-1):
    if ops[i] != ' ':
        ops_clean.append(ops[i])
    elif ops[i+1] != ' ':
        ops_clean.append(',')
    else:
        ops_clean.append('_')
ops_clean.append('_')
ops_clean = ''.join(ops_clean)
dig_counts = [op.count('_')+1 for op in ops_clean.split(',')]
ops_clean, dig_counts
```

```python
lol = []
for line in lines[:-1]:
    line_clean = ''.join([l if ops_clean[i] != ',' else ',' for i,l in enumerate(line)])
    lol.append([l[::-1] for l in line_clean.split(',')])

lol
```

```python
numbers = []
for ind in range(len(dig_counts)):
    col = np.array(lol).T[ind]
    nums = []
    for dig in range(dig_counts[ind]):
        nums.append(int(''.join([c[dig] for c in col])))
    # print(nums)
    numbers.append(nums)
# numbers = np.array(numbers)
```
